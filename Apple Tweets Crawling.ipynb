```python
from twitterscraper import query_tweets
import datetime as dt
import pandas as pd
```

```python
NameOfCompanies = ["@Apple", "iphones", "itunes","airpods","applesupport","macbook"]
companies = []
```

```python
i=0
while i < len(NameOfCompanies):   
    begin_date = dt.date(2019,10,1)
    end_date = dt.date(2020,2,1)
    limit = 0
    lang = "english"
    CompanyTweets = query_tweets(NameOfCompanies[i], limit=limit, begindate=begin_date, enddate=end_date, lang=lang)
    companies.append(CompanyTweets)
    i= i+1
else:
    print(NameOfCompanies[0])
    print(NameOfCompanies[1])
    print(NameOfCompanies[2])
    print(NameOfCompanies[3])
    print(NameOfCompanies[4])
    print(NameOfCompanies[5])
    print("Finally finished!")
```

```python
dfNames = ["@Apple", "iphones", "itunes","airpods","applesupport","macbook"]
```

```python
i=0
for company in companies:
    dfNames[i] = pd.DataFrame(t.__dict__ for t in  company)
    i = i+1
```

```python
Apple = dfNames[0]
iphones = dfNames[1]
itunes = dfNames[2]
airpods = dfNames[3]
applesupport = dfNames[4]
macbook = dfNames[5]
```

```python
AppleTweets = Apple.append([iphones,itunes,airpods,applesupport,macbook])
```

```python
AppleTweets.reset_index(inplace=True, drop=True)
```
